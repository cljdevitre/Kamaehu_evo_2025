## Load data

```python
Nov1=pd.read_excel(r'Nov1st_WDSEDS_S.xlsx')
Nov2122=pd.read_excel(r'Nov2122_WDSEDS_S.xlsx')
Nov1 = Nov1.loc[:, ~Nov1.columns.str.contains('^Unnamed')]
Nov2122 = Nov2122.loc[:, ~Nov2122.columns.str.contains('^Unnamed')]
Nov1['AnalysisDate_EDS']='11/01/2024'
Nov1['AnalysisDate_WDS']='11/01/2024'
Nov2122['AnalysisDate_EDS']='11/21/2024'
Nov2122['AnalysisDate_WDS']='11/21/2024'

combo=pd.concat([Nov1,Nov2122],axis=0)
combo
```

## Compare EDS and WDS

```python
df_combined_WDS_EDS[df_combined_WDS_EDS['Sample Name_WDS'].str.contains("JDF_g2_thu_noon_4_WDS_endrun")]
```

```python
ele='S'
oxide='SO3'

df_combined_WDS_EDS = combo.copy()

old=(df_combined_WDS_EDS['AnalysisDate_EDS']=='11/01/2024')
new=(df_combined_WDS_EDS['AnalysisDate_EDS']=='11/21/2024')
standards=(df_combined_WDS_EDS['Sample Name_WDS'].str.contains('JDF'))
samples=~(df_combined_WDS_EDS['Sample Name_WDS'].str.contains('JDF'))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Plot the error bars


ax1.errorbar(df_combined_WDS_EDS[f'{oxide}_WDS'][samples&new], df_combined_WDS_EDS[f'{oxide}_EDS'][samples&new], 
             xerr=df_combined_WDS_EDS[f'Oxide % Sigma_{oxide}_WDS'][samples&new], 
             yerr=df_combined_WDS_EDS[f'Oxide % Sigma_{oxide}_EDS'][samples&new],
             fmt='d', ecolor='grey', elinewidth=0.8, mfc='orange', ms=5, mec='k', capsize=3,label='Samples new holder')
ax1.errorbar(df_combined_WDS_EDS[f'{oxide}_WDS'][samples&old], df_combined_WDS_EDS[f'{oxide}_EDS'][samples&old], 
             xerr=df_combined_WDS_EDS[f'Oxide % Sigma_{oxide}_WDS'][samples&old], 
             yerr=df_combined_WDS_EDS[f'Oxide % Sigma_{oxide}_EDS'][samples&old],
             fmt='d', ecolor='grey', elinewidth=0.8, mfc='lightgrey', ms=5, mec='k', capsize=3, label='Samples old holder')

ax2.errorbar(df_combined_WDS_EDS[f'{oxide}_WDS'][standards&new], df_combined_WDS_EDS[f'{oxide}_EDS'][standards&new], 
             xerr=df_combined_WDS_EDS[f'Oxide % Sigma_{oxide}_WDS'][standards&new], 
             yerr=df_combined_WDS_EDS[f'Oxide % Sigma_{oxide}_EDS'][standards&new],
             fmt='d', ecolor='grey', elinewidth=0.8, mfc='orange', ms=5, mec='k', capsize=3, label='Standards new holder')

ax2.errorbar(df_combined_WDS_EDS[f'{oxide}_WDS'][standards&old], df_combined_WDS_EDS[f'{oxide}_EDS'][standards&old], 
             xerr=df_combined_WDS_EDS[f'Oxide % Sigma_{oxide}_WDS'][standards&old], 
             yerr=df_combined_WDS_EDS[f'Oxide % Sigma_{oxide}_EDS'][standards&old],
             fmt='d', ecolor='grey', elinewidth=0.8, mfc='lightgrey', ms=5, mec='k', capsize=3, label='Standards old holder')


only_50nA=(df_combined_WDS_EDS['Sample Name_WDS'].str.contains("JDF_g2_thu_noon_4_WDS_endrun"))&(df_combined_WDS_EDS['Sample Name_WDS'].str.contains("50nA"))
comp_100vs50=(df_combined_WDS_EDS['Sample Name_WDS'].str.contains("JDF_g2_thu_noon_4_WDS_endrun"))&~(df_combined_WDS_EDS['Sample Name_WDS'].str.contains("50nA"))

ax2.errorbar(df_combined_WDS_EDS[f'{oxide}_WDS'][only_50nA], df_combined_WDS_EDS[f'{oxide}_EDS'][only_50nA], 
             xerr=df_combined_WDS_EDS[f'Oxide % Sigma_{oxide}_WDS'][only_50nA], 
             yerr=df_combined_WDS_EDS[f'Oxide % Sigma_{oxide}_EDS'][only_50nA],
             fmt='d', ecolor='grey', elinewidth=0.8, mfc='cyan', ms=5, mec='k', capsize=3, label='50nA')

ax2.errorbar(df_combined_WDS_EDS[f'{oxide}_WDS'][comp_100vs50], df_combined_WDS_EDS[f'{oxide}_EDS'][comp_100vs50], 
             xerr=df_combined_WDS_EDS[f'Oxide % Sigma_{oxide}_WDS'][comp_100vs50], 
             yerr=df_combined_WDS_EDS[f'Oxide % Sigma_{oxide}_EDS'][comp_100vs50],
             fmt='d', ecolor='grey', elinewidth=0.8, mfc='blue', ms=5, mec='k', capsize=3, label='100nA')

combo[combo['Sample Name_WDS'].str.contains("50")]
combo[combo['Sample Name_WDS'].str.contains("JDF_g2_thu_noon_4_WDS_endrun1")]
# Plot the reference lines
max=np.max(df_combined_WDS_EDS[f'{oxide}_WDS'])
min=np.min(df_combined_WDS_EDS[f'{oxide}_WDS'])

ax1.plot([min-0.02, max], [min-0.02, max], '-k',label="1:1")
ax1.plot([min-0.02, max], [min-0.02, max* 1.1], ':k',label='+-10%')
ax1.plot([min-0.02, max], [min-0.02, max * 0.9], ':k')

# Set the labels dynamically
ax1.set_xlabel(f'WDS {oxide} (wt%)')
ax1.set_ylabel(f'EDS {oxide} (wt%)')

ax2.plot([min-0.02, max], [min-0.02, max], '-k',label="1:1")
ax2.plot([min-0.02, max], [min-0.02, max* 1.1], ':k',label='+-10%')
ax2.plot([min-0.02, max], [min-0.02, max * 0.9], ':k')


max_standards=np.max(df_combined_WDS_EDS[f'{oxide}_WDS'][standards])
min_standards=np.min(df_combined_WDS_EDS[f'{oxide}_WDS'][standards])

ax2.set_xlim([min_standards-0.05, max_standards+0.07])
ax2.set_ylim([min_standards-0.05, max_standards+0.07])

ax1.set_xlim([min-0.02, max+0.05])
ax1.set_ylim([min-0.02, max+0.05])
# Set the labels dynamically
ax2.set_xlabel(f'WDS {oxide} (wt%)')
ax2.set_ylabel(f'EDS {oxide} (wt%)')

ax2.legend()
ax1.legend()
```
